```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,cross_val_score,cross_validate,GridSearchCV,RandomizedSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
from sklearn.preprocessing import StandardScaler
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=230)
scalar = StandardScaler()
x_train1 = scalar.fit_transform(x_train)
x_test1 = scalar.transform(x_test)
x1 = scalar.fit_transform(x)
```

```python
dt1 = DecisionTreeClassifier(max_depth=9,max_features=0.8,class_weight={0:3,1:6})
dt1.fit(x_train,y_train)
y_pred1 = dt1.predict(x_test)
y_proba1 = dt1.predict_proba(x_test)[:,1]
print('Accuracy :',np.round(accuracy_score(y_test,y_pred1),3))
print('Precision :',np.round(precision_score(y_test,y_pred1),3))
print("Recall :",np.round(recall_score(y_test,y_pred1),3))
print('f1_score :',np.round(f1_score(y_test,y_pred1),3))
print('roc_auc :',np.round(roc_auc_score(y_test,y_proba1),3))
```

```python
splitter = ['best','random']
criterian = ['gini','entropy','log_loss']
max_features = ['sqrt','log2',None,0.8,0.7,0.6,0.5]
max_depth = [3,5,7,9,11,13,15]
class_weight = [1,2,3,4,5]
ccp_alpha = [0.0, 0.001, 0.01]
accuracy = []
precision = []
recalls = []
f1_scores = []
roc_auc = []
splitter1 = []
max_depth1 = []
criterian1 = []
max_features1 = []
class_weight_0 = []
class_weight_1 = []
ccp_alpha1 = []
for depth in max_depth:
    for criteria in criterian:
        for split in splitter:
            for feature in max_features:
                for i in class_weight:
                    for alpha in ccp_alpha:
                        dt = DecisionTreeClassifier(criterion=criteria,max_depth=depth,splitter=split,
                                                    max_features=feature,class_weight={0:i,1:2*i},ccp_alpha=alpha)
                        dt.fit(x_train1,y_train)
                        y_pred = dt.predict(x_test1)
                        y_proba = dt.predict_proba(x_test1)[:,1]
                        accu = np.round(accuracy_score(y_test,y_pred),3)
                        preci = np.round(precision_score(y_test,y_pred),3)
                        recall = np.round(recall_score(y_test,y_pred),3)
                        f1 = np.round(f1_score(y_test,y_pred),3)
                        roc = np.round(roc_auc_score(y_test,y_proba),3)
                        accuracy.append(accu)
                        precision.append(preci)
                        recalls.append(recall)
                        f1_scores.append(f1)
                        roc_auc.append(roc)
                        max_depth1.append(depth)
                        criterian1.append(criteria)
                        splitter1.append(split)
                        max_features1.append(feature)
                        class_weight_0.append(i)
                        class_weight_1.append(2*i)
                        ccp_alpha1.append(alpha)
                    
```

```python
temp = pd.DataFrame({
        'max_depth':max_depth1,
        'criterian':criterian1,
        'splitter':splitter1,
        'max_features':max_features1,
        'class_weight_o':class_weight_0,
        'class_weight_1':class_weight_1,
        'ccp_alpha':ccp_alpha1,
        'accuracy':accuracy,
        'precision':precision,
        'recall':recalls,
        'f1_score':f1_scores,
        'roc_auc':roc_auc,
    })
```

```python
param_distributions = {
    'splitter': ['best', 'random'],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_features': ['sqrt', 'log2', None, 0.5,0.8],
    'max_depth': [3, 5, 7, 10,12, None],
    'ccp_alpha': [0.0, 0.001, 0.01],
    'min_samples_leaf': [1, 2, 5],
    'min_samples_split': [2, 5, 10]
}

scoring = {
    'acc': 'accuracy',
    'prec': 'precision',
    'rec': 'recall',
    'f1': 'f1',
    'auc': 'roc_auc'
}

random_grid = RandomizedSearchCV(
    DecisionTreeClassifier(),
    param_distributions,
    n_jobs=-1,
    scoring=scoring,
    refit=False,
    cv = 5
)

random_grid.fit(x_train,y_train)
temp_df = pd.DataFrame(random_grid.cv_results_)
```
