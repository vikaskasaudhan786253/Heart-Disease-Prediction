```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,roc_auc_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('display.max_colWidth',None)
```

```python
df = pd.read_csv('train_cleaned.csv')
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=230)
scalar = StandardScaler()
x_train1 = scalar.fit_transform(x_train)
x_test1 = scalar.transform(x_test)
x1 = scalar.fit_transform(x)
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
y_pred1 = svm.predict(x_test)
y_proba1 = svm.predict_proba(x_test)[:,1]
print('Accuracy :',np.round(accuracy_score(y_test,y_pred1),3))
print('Precision :',np.round(precision_score(y_test,y_pred1),3))
print("Recall :",np.round(recall_score(y_test,y_pred1),3))
print('f1_score :',np.round(f1_score(y_test,y_pred1),3))
print('roc_auc :',np.round(roc_auc_score(y_test,y_proba1),3))
```

```python
knn = KNeighborsClassifier(n_jobs=-1,weights='distance')
knn.fit(x_train,y_train)
y_pred1 = knn.predict(x_test)
y_proba1 = knn.predict_proba(x_test)[:,1]
print('Accuracy :',np.round(accuracy_score(y_test,y_pred1),3))
print('Precision :',np.round(precision_score(y_test,y_pred1),3))
print("Recall :",np.round(recall_score(y_test,y_pred1),3))
print('f1_score :',np.round(f1_score(y_test,y_pred1),3))
print('roc_auc :',np.round(roc_auc_score(y_test,y_proba1),3))
```

```python
knn1 = KNeighborsClassifier(n_jobs=-1,n_neighbors=8,weights='distance')
knn1.fit(x_train,y_train)
y_pred1 = knn1.predict(x_test)
y_proba1 = knn1.predict_proba(x_test)[:,1]
print('Accuracy :',np.round(accuracy_score(y_test,y_pred1),3))
print('Precision :',np.round(precision_score(y_test,y_pred1),3))
print("Recall :",np.round(recall_score(y_test,y_pred1),3))
print('f1_score :',np.round(f1_score(y_test,y_pred1),3))
print('roc_auc :',np.round(roc_auc_score(y_test,y_proba1),3))
```

```python
knn2 = KNeighborsClassifier(n_jobs=-1,n_neighbors=10,weights='distance')
knn2.fit(x_train,y_train)
y_pred1 = knn2.predict(x_test)
y_proba1 = knn2.predict_proba(x_test)[:,1]
print('Accuracy :',np.round(accuracy_score(y_test,y_pred1),3))
print('Precision :',np.round(precision_score(y_test,y_pred1),3))
print("Recall :",np.round(recall_score(y_test,y_pred1),3))
print('f1_score :',np.round(f1_score(y_test,y_pred1),3))
print('roc_auc :',np.round(roc_auc_score(y_test,y_proba1),3))
```

```python
accuracy , precisions , recalls, f1_scores, roc_auc = [],[],[],[],[]
n_neighbors = [5,10,15]
weights = ['uniform','distance']
algorithm = ['auto','ball_tree','kd_tree','brute']
p = [1,2,3]
n_neighbors1 , weights1, algorithms1,p1 = [],[],[],[]
for algo in algorithm:
    for neighbor in n_neighbors:
        for weight in weights:
            for i in p:
                knn = KNeighborsClassifier(n_jobs=-1,n_neighbors=neighbor,
                                           weights=weight,algorithm=algo,p=i)
                knn.fit(x_train,y_train)
                y_pred1 = knn.predict(x_test)
                y_proba1 = knn.predict_proba(x_test)[:,1]
                accu = np.round(accuracy_score(y_test,y_pred1),3)
                prec = np.round(precision_score(y_test,y_pred1),3)
                recall = np.round(recall_score(y_test,y_pred1),3)
                f1 = np.round(f1_score(y_test,y_pred1),3)
                roc = np.round(roc_auc_score(y_test,y_proba1),3)
                accuracy.append(accu)
                precisions.append(prec)
                recalls.append(recall)
                f1_scores.append(f1)
                roc_auc.append(roc)
                n_neighbors1.append(neighbor)
                weights1.append(weight)
                algorithms1.append(algo)
                p1.append(i)
```

```python
df = pd.DataFrame({
    'accuracy':accuracy,
    'precisions':precisions,
    'recalls':recalls,
    'f1_scores':f1_scores,
    'roc_auc':roc_auc,
    'n_neighbors':n_neighbors1,
    'weights':weights1,
    'algorithms':algorithms1,
    'p':p1
})
```

```python
df.to_csv('knn_all_combi.csv',index=False)
```
